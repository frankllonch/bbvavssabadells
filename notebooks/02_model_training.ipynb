```python
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from src.model_rnn import create_rnn

tickers = ['BBVA.MC', 'SAB.MC']  # your keys
sequence_length = 30

for ticker in tickers:
    print(f"\nğŸ”¥ Training RNN for {ticker}...")
    
    # Load preprocessed data
    X_train = np.load(f"/Users/frankllonch/Desktop/quattroporte/aprendizado de mÃ¡quina/bbvavssabadells/data/processed/X_train_{ticker}.npy")
    y_train = np.load(f"/Users/frankllonch/Desktop/quattroporte/aprendizado de mÃ¡quina/bbvavssabadells/data/processed/y_train_{ticker}.npy")
    
    # Create model
    model = create_rnn(input_shape=(X_train.shape[1], X_train.shape[2]))
    
    # Callbacks
    checkpoint = ModelCheckpoint(f"models/rnn_{ticker}.h5", save_best_only=True, monitor='loss', verbose=1)
    early_stop = EarlyStopping(monitor='loss', patience=20, restore_best_weights=True)
    
    # Train
    history = model.fit(
        X_train, y_train,
        epochs=800,
        batch_size=32,
        callbacks=[checkpoint, early_stop],
        verbose=1
    )
    
    print(f"âœ… Finished training {ticker}")
```
